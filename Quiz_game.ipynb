```python
import random

class QuizGame:
    def __init__(self):
        self.questions_pool = []
        self.selected_questions = []
        self.score = 0
        self.total_questions = 0

    def welcome_message(self):
        print("\nWelcome to the Quiz Game!")
        print("Rules:")
        print("1. You'll be asked multiple-choice or fill-in-the-blank questions.")
        print("2. You get 1 point for each correct answer.")
        print("3. Your final score will be calculated at the end.")
        print("4. Let's see how well you know your stuff!\n")
        print("5. The System is Case sensitive")
        print("Good luck!\n")

    def load_questions(self):

        self.questions_pool = [
            {
        'type': 'multiple-choice',
        'question': "What is the capital of France?",
        'choices': ['A) Berlin', 'B) Paris', 'C) Madrid', 'D) Rome'],
        'answer': 'B'

    },
    {
        'type': 'multiple-choice',
        'question': "What is the capital of Japan?",
        'choices': ['A) Beijing', 'B) Tokyo', 'C) Seoul', 'D) Bangkok'],
        'answer': 'B'
    },
    {
        'type': 'multiple-choice',
        'question': "Which element has the chemical symbol 'O'?",
        'choices': ['A) Oxygen', 'B) Osmium', 'C) Oganesson', 'D) Opium'],
        'answer': 'A'
    },
    {
        'type': 'multiple-choice',
        'question': "Which is the largest planet in our solar system?",
        'choices': ['A) Earth', 'B) Venus', 'C) Jupiter', 'D) Saturn'],
        'answer': 'C'
    },
    {
        'type': 'multiple-choice',
        'question': "Who painted the Mona Lisa?",
        'choices': ['A) Michelangelo', 'B) Leonardo da Vinci', 'C) Pablo Picasso', 'D) Vincent van Gogh'],
        'answer': 'B'
    },
    {
        'type': 'multiple-choice',
        'question': "Which country is known as the Land of the Rising Sun?",
        'choices': ['A) China', 'B) Japan', 'C) India', 'D) Thailand'],
        'answer': 'B'
    },
    {
        'type': 'multiple-choice',
        'question': "What is the square root of 64?",
        'choices': ['A) 6', 'B) 8', 'C) 7', 'D) 9'],
        'answer': 'B'
    },
    {
        'type': 'multiple-choice',
        'question': "Which language is the most spoken worldwide?",
        'choices': ['A) Spanish', 'B) English', 'C) Mandarin Chinese', 'D) Hindi'],
        'answer': 'C'
    },
    {
        'type': 'multiple-choice',
        'question': "Which country won the FIFA World Cup in 2018?",
        'choices': ['A) Brazil', 'B) Germany', 'C) France', 'D) Argentina'],
        'answer': 'C'
    },
    {
        'type': 'multiple-choice',
        'question': "Which metal is heavier, silver or gold?",
        'choices': ['A) Silver', 'B) Gold', 'C) Both are equal', 'D) Neither'],
        'answer': 'B'
    },
    {
        'type': 'multiple-choice',
        'question': "Which planet is closest to the sun?",
        'choices': ['A) Venus', 'B) Mars', 'C) Mercury', 'D) Earth'],
        'answer': 'C'
    },

    {
        'type': 'multiple-choice',
        'question': "What is the capital of France?",
        'choices': ['A) Berlin', 'B) Madrid', 'C) Paris', 'D) Rome'],
        'answer': 'C'
     },
     {
        'type': 'multiple-choice',
        'question': "Which planet is known as the Red Planet?",
        'choices': ['A) Earth', 'B) Mars', 'C) Jupiter', 'D) Saturn'],
        'answer': 'B'
    },
    {
        'type': 'fill-in-the-blank',
        'question': "The chemical symbol for water is ____.",
        'answer': 'H2O'
     },
    {
        'type': 'fill-in-the-blank',
        'question': "Who wrote 'Hamlet'? ____.",
        'answer': 'Shakespeare'
     },
     {
        'type': 'multiple-choice',
        'question': "What is the largest ocean on Earth?",
        'choices': ['A) Atlantic Ocean', 'B) Indian Ocean', 'C) Arctic Ocean', 'D) Pacific Ocean'],
        'answer': 'D'
    },
    {
        'type': 'fill-in-the-blank',
        'question': "The process of converting water into vapor is called ____.",
        'answer': 'evaporation'
    },
    {
        'type': 'multiple-choice',
        'question': "How many continents are there on Earth?",
        'choices': ['A) 5', 'B) 6', 'C) 7', 'D) 8'],
        'answer': 'C'
    },
    {
        'type': 'fill-in-the-blank',
        'question': "Light travels at a speed of approximately ____ kilometers per second.",
        'answer': '300000'
    },
    {
        'type': 'multiple-choice',
        'question': "Who developed the theory of relativity?",
        'choices': ['A) Isaac Newton', 'B) Galileo Galilei', 'C) Nikola Tesla', 'D) Albert Einstein'],
        'answer': 'D'
    },
    {
        'type': 'fill-in-the-blank',
        'question': "The largest mammal on Earth is the ____.",
        'answer': 'blue whale'
    }
 ]

    def select_random_questions(self, num_questions):

        random.shuffle(self.questions_pool)
        self.selected_questions = self.questions_pool[:num_questions]
        self.total_questions = len(self.selected_questions)

    def present_question(self, question):
        if question['type'] == 'multiple-choice':
            print(question['question'])
            for choice in question['choices']:
                print(choice)
            user_answer = input("Choose the correct answer (A, B, C, or D): ").upper()
        else:  # fill-in-the-blank
            print(question['question'])
            user_answer = input("Your answer: ").strip()

        return user_answer

    def evaluate_answer(self, user_answer, correct_answer):
        if user_answer == correct_answer:
            print("Correct!\n")
            self.score += 1
        else:
            print(f"Incorrect! The correct answer was {correct_answer}.\n")

    def display_final_results(self):
        print(f"\nFinal Score: {self.score}/{self.total_questions}")
        if self.score == self.total_questions:
            print("Excellent! You got everything right!")
        elif self.score >= self.total_questions / 2:
            print("Good job! You did well!")
        else:
            print("Better luck next time!")

    def play_again(self):
        play_again_choice = input("\nDo you want to play again? (yes/no): ").strip().lower()
        return play_again_choice == 'yes'

    def play_game(self, num_questions):
        self.welcome_message()
        self.load_questions()
        self.select_random_questions(num_questions)
        self.score = 0

        for question in self.selected_questions:
            user_answer = self.present_question(question)
            self.evaluate_answer(user_answer, question['answer'])

        self.display_final_results()


def main():
    game = QuizGame()
    num_questions = int(input("How many questions would you like to answer? "))

    while True:
        game.play_game(num_questions)
        if not game.play_again():
            print("Thanks for playing! Goodbye!")
            break

if __name__ == "__main__":
    main()
```
