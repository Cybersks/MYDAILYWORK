```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

questions = [
    {
        "question": "What is the capital of France?",
        "options": ["Berlin", "Paris", "Madrid", "Rome"],
        "answer": "Paris"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["Earth", "Mars", "Jupiter", "Saturn"],
        "answer": "Mars"
    },
    {
        "question": "Who wrote 'Hamlet'?",
        "options": ["Charles Dickens", "J.K. Rowling", "William Shakespeare", "Leo Tolstoy"],
        "answer": "William Shakespeare"
    },
    {
        "question": "What is the largest ocean on Earth?",
        "options": ["Atlantic", "Indian", "Arctic", "Pacific"],
        "answer": "Pacific"
    },
    {
        "question": "What is the chemical symbol for gold?",
        "options": ["Au", "Ag", "Fe", "Pb"],
        "answer": "Au"
    }
]

random.shuffle(questions)

current_question_index = 0
score = 0

def display_question():
    global current_question_index

    clear_output()

    if current_question_index >= len(questions):
        display(widgets.HTML(f"<h2>Quiz Completed! Your score is {score}/{len(questions)}.</h2>"))
        return

    question_data = questions[current_question_index]

    question_label = widgets.HTML(f"<h3>{question_data['question']}</h3>")
    display(question_label)

    options = question_data["options"]
    radio_buttons = widgets.RadioButtons(
        options=options,
        description='',
        disabled=False
    )
    display(radio_buttons)


    submit_button = widgets.Button(description="Submit")
    display(submit_button)


    def on_submit(b):
        global current_question_index, score
        selected_option = radio_buttons.value

        if selected_option == question_data["answer"]:
            score += 1

        current_question_index += 1
        display_question()

    submit_button.on_click(on_submit)


display_question()
```
